```python
### Book Inbalance
```

```python
def book_volume_inbalance(sym, idx):
    ge_bid_price = bid_price[sym][idx] >= bid_price[sym][idx-1]
    le_bid_price = bid_price[sym][idx] <= bid_price[sym][idx-1]
    ge_ask_price = ask_price[sym][idx] >= ask_price[sym][idx-1]
    le_ask_price = ask_price[sym][idx] <= ask_price[sym][idx-1]
    en = ge_bid_price * bid_size[sym][idx] - le_bid_price * bid_size[sym][idx-1] + \
         ge_ask_price * ask_size[sym][idx-1] - le_ask_price * ask_size[sym][idx]
    return en
```

```python
book_inbalances = dict()
for sym in symbols:
    en = [book_volume_inbalance(sym, i) for i in range(1, len(bid_price[sym]))]
    book_inbalances[sym] = en
```

```python
### Book Return
```

```python
book_returns = dict()
```

```python
def wap(bid_price, ask_price, bid_size, ask_size):
    return (bid_price * ask_size + ask_price * bid_size) / (bid_size + ask_size)
```

```python
def log_return(price1, price2):
    return math.log(price2 / price1)
```

```python
for sym in symbols:
    waps = [wap(bp, ap, bz, az) for bp, ap, bz, az in zip(bid_price[sym], ask_price[sym], bid_size[sym], ask_size[sym])]
    lrs = [log_return(p1, p2) for p1, p2 in zip(waps[:-1], waps[1:])]
    book_returns[sym] = lrs
```

```python
### Book Realized Volatility
```

```python
book_volatilities = dict()
```

```python
def realized_volatility(best_bid_prices, best_ask_prices, best_bid_sizes, best_ask_sizes):
    prices = [wap(bp, ap, bz, az) for bp, ap, bz, az in zip(best_bid_prices, best_ask_prices, best_bid_sizes, best_ask_sizes)]
    returns = [log_return(p1, p2) for p1, p2 in zip(prices[:-1], prices[1:])]
    #TODO: should i divide by the size?
    volatility = math.sqrt(sum([r**2 for r in returns]))
    return volatility
```

```python
for sym in symbols:
    book_volatilities[sym] = realized_volatility(bid_price[sym], ask_price[sym], bid_size[sym], ask_size[sym])
```

```python
### Trade Prices and Sizes per second
```

```python
datetime_format = "%Y-%m-%dT%H:%M:%S.%f %Z"
```

```python
min_time = datetime.strptime(trade['sequence_time'].min(), datetime_format)
max_time = datetime.strptime(trade['sequence_time'].max(), datetime_format)
print("{},{}".format(min_time, max_time))
```

```python
trade_sizes = dict()
trade_avg_prices = dict()
```

```python
for symbol in symbols:
    start_time = min_time
    end_time = max_time
    size_lst = []
    avg_price_lst = []
    while start_time < end_time:
        next_time = start_time + timedelta(seconds=1)
        rows = trade.query('pid == "{}" and time >= "{}" and time < "{}"'.format(symbol, start_time, next_time))
        trade_size = rows['size'].tolist()
        trade_price = rows['price'].tolist()
        size_lst.append(sum(trade_size))
        if len(trade_price) == 0:
            avg_price_lst.append(float("nan"))
        else:
            avg_price_lst.append(sum([p * z for p, z in zip(trade_price, trade_size)]) / sum(trade_size))
        start_time = next_time
    trade_sizes[symbol] = size_lst
    trade_avg_prices[symbol] = avg_price_lst
```
