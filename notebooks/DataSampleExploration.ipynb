### Financial Concepts

##### bid ask spread is a percentage of value between bid and ask, usually the best bid and the best ask
##### note ask is always biggeer than bid (ask > bid) in a order book

```python
# (ask - bid) / bid
def bid_ask_spread1(best_bid, best_ask):
    return best_ask / best_bid - 1.
```

```python
# (ask - bid) / ask
def bid_ask_spread2(best_bid, best_ask):
    return 1. - best_bid / best_ask
```

##### weighted average price, a fair book based valuation

```python
def wap(bid_price, ask_price, bid_size, ask_size):
    return (bid_price * ask_size + ask_price * bid_size) / (bid_size + ask_size)
```

##### log return is better because lr_t1 + lr_t2 = lr_(t1+t2)

```python
def log_return(price1, price2):
    return log(price2 / price1)
```

##### formula is for realized volatility over any time period (since 1 year is irrelevant)

##### typical price is the average of high, low and closing price of the candlestick

```python
def typical_price(high_price, low_price, close_price):
    return (high_price + low_price + close_price) / 3.
```

##### volume weighted average price (VWAP)
##### is computed by typical_price * running_volume / total_volume
##### where running volume is the volume for the candlestick period, and
##### total volume is the volume of the day, or some fixed period

```python
def vwap(typical_price, volume, cum_volume):
    return typical_price * volume / cum_volume
```
