```python
import pandas as pd
```

```python
data_dir = '../data/'
```

```python
files = [
    data_dir + 'etl_s1_1672118400_1672803840.parquet',
    #data_dir + 'etl_s1_1671504000_1627118400.parquet',
]
```

```python
data = pd.read_parquet(files[0])
```

```python
columns = data.columns
```

```python
rcol960 = []
rcol1440 = []
rcol1920 = []
for column in columns:
    if 'book_return_960sum' in column:
        rcol960.append(column)
    elif 'book_return_1440sum' in column:
        rcol1440.append(column)
    elif 'book_return_1920sum' in column:
        rcol1920.append(column)
```

```python
data[rcol960].describe()
```

```python
data.shape[0]
```

```python
for col in rcol960:
    profit = sum(data[col] > 0.0065)
    print("{}: rows: {} pct: {}".format(col, profit, profit / data.shape[0]))
```

```python
data[rcol1440].describe()
```

```python
for col in rcol1440:
    profit = sum(data[col] > 0.0065)
    print("{}: rows: {} pct: {}".format(col, profit, profit / data.shape[0]))
```

```python
for col in rcol1920:
    profit = sum(data[col] > 0.0065)
    print("{}: rows: {} pct: {}".format(col, profit, profit / data.shape[0]))
```
