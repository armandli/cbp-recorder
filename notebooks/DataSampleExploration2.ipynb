```python
### looking at whether any sequence is missing
```

```python
timestamp_format = '%Y-%m-%dT%H:%M:%S.%f %Z'
```

```python
def sequence_epoch(row):
    utc = pytz.timezone('UTC')
    return int(datetime.strptime(row['sequence_time'], timestamp_format).timestamp())
```

```python
book10_data['sequence_epoch'] = book10_data.apply(lambda row: sequence_epoch(row), axis=1)
book10_data.set_index('sequence_epoch')
```

```python
(book10_data.index[1:]-book10_data.index[:-1]).value_counts().head()
```
